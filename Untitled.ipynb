## 19. Оптимизация. Задача упаковки кругов

Рассмотрим задачу упаковки в единичный круг заданного числа $n$ равных друг другу меньших кругов радиуса $r$, так, чтобы радиус $r$ был максимальным [S. Kravitz, Packing cylinders into cylindrical containers, Math. Mag. 40 (1967), p. 65–71].

Перейдем к математической постановке задачи.
Обозначим через $(x_j, y_j)$ – координаты центра $j$-го круга.
Условия того, чтобы эти круги помещались в единичном круге $x^2+y^2\le 1$,
принимает вид:
$$
  x_j^2 + y_j^2 \le (1 - r)^2 \qquad (j=1,2,\dots,n).
$$
Условие того, что круги не пересекаются, имеет вид:
$$
  (x_i - x_j)^2 + (y_i - y_j)^2 \ge 4 r^2 \qquad (1 \le i < j \le n).
$$
Также необходимо учитывать ограничения
$$
  0 \le r \le 1.
$$
Таким образом, задача заключается в максимизции $r$ при указанных выше ограничениях.

Задача содержит $2n+1$ переменных (неизвестных) и $n + n(n-1)/2 + 2$
ограничений-неравенств.

Например, при $n=3$ получаем задачу
$$
r \to \max 
$$
при ограничениях
$$
\left\{
  \begin{array}{l}
     0 \le r \le 1, \\
     x_1^2 + y_1^2 \le (1 - r)^2, \\
     x_2^2 + y_2^2 \le (1 - r)^2, \\
     x_3^2 + y_3^2 \le (1 - r)^2, \\
     (x_1 - x_2)^2 + (y_1 - y_2)^2 \ge 4 r^2, \\
     (x_1 - x_3)^2 + (y_1 - y_3)^2 \ge 4 r^2, \\
     (x_2 - x_3)^2 + (y_2 - y_3)^2 \ge 4 r^2. \\
  \end{array}
\right.   
$$

Написать функцию, принимающую на вход $n$ и находящую оптимальную упаковку.
Функция должна использовать функцию `scipy.optimize.minimize`.

 Функция circ  ограничения для кругов (все значения массива "d" должны быть неотрицательными)

```python
def circ(x):
    d = []
    for i in range(1, 2*n, 2):
        d = np.append(d, (x[0]-1)**2 - x[i]**2 - x[i+1]**2) ## добавление в конец массива d
    for i in range(1, 2*n, 2):
        for j in range(i+2, 2*n + 1, 2):
            d = np.append(d, (x[i] - x[j])**2 + (x[i+1] - x[j+1])**2 - 4)
    d = np.append(d, x[0] - 1)
    d = np.append(d, n - x[0])
    return d
```

Функция main создаёт массив из 2n+1 элементов, где a[0] элемент - радиус R, (a[i],a[i+1]), i=1,3,...2n+1 - координаты маленьких кругов.
После функции mininize идёт приведение к первоначальному условию.
Всё это выполняется в цикле,потому работает не сразу 

```python
def main(n):
    t = 0;
    while (t < primer[n-1] - 0.001):
        a = [random.uniform(-n,n) for i in range(2*n+1)]
        a[0] = n
        res = minimize(lambda x: x[0], a, constraints=({'type':'ineq','fun': circ}))
        x = res.x / res.x[0]
        x[0] /= res.x[0]
        t = x[0]
    paint(n, x)## вызов  отображения
    print("Разница с таблицей взятой с сайта :", abs(t-primer[n-1]))
    return
```
